```python
from sklearn import datasets
import matplotlib.pyplot as plt
```

# digits

```python
digits = datasets.load_digits()

# データの属性リスト表示
# データのサイズ表示
print('\ndir: {}'.format(dir(digits)))
print('\ndata.shape: {}'.format(digits.data.shape))
print('\nimage.shape: {}'.format(digits.images.shape)) 

# 1枚目のベクトルデータ表示
# 1枚目の行列データ表示
# 1枚目のラベル表示
# 1枚目のラベル名表示
print('\ndata[0]: {}'.format(digits.data[0]))
print('\nimages[0]: {}'.format(digits.images[0]))
print('\ntarget[0]: {}'.format(digits.target[0]))
print('\ntarget_names[0]: {}'.format(digits.target_names[0]))

# 1枚目の画像表示
plt.imshow(digits.images[0]);
plt.show()

# データの説明表示
print('\nDESCR: {}'.format(digits.DESCR))
```

# sample_image

```python
# サンプル画像の読み込み china.jpg/flower.jpg
china = datasets.load_sample_image('china.jpg') 

# numpyデータ構造表示
# データ構造表示
# データサイズ表示
print('type: {}'.format(type(china)))
print('dtype: {}'.format(china.dtype))
print('shape: {}'.format(china.shape ))

# 画像表示
plt.imshow(china)
plt.show()
```

```python
# サンプル画像の読み込み flower.jpg
flower = datasets.load_sample_image('flower.jpg') 

# numpyデータ構造表示
# データ構造表示
# データサイズ表示
print('type: {}'.format(type(flower)))
print('dtype: {}'.format(flower.dtype))
print('shape: {}'.format(flower.shape ))

# 画像表示
plt.imshow(flower)
plt.show()
```

```python
# 画像データセットの読み込み
dataset = datasets.load_sample_images()

# オブジェクト型を確認
# 画像枚数を確認
print(type(dataset))
print(len(dataset.images))

# 1枚目の画像
img0 = dataset.images[0] 
print(img0.shape)
print(img0.dtype)
plt.imshow(img0)
plt.show()

# 2枚目の画像
img1 = dataset.images[1]
print(img1.shape)
print(img1.dtype)
plt.imshow(img1)
plt.show()
```
