```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.array([[4,7],[8,6.5],[6,3],[2,1],[1,2],[3,5],[2.5,6],[1.5,9]])

plt.figure(figsize=(6,6))
plt.scatter(*X.T)
plt.axis([0, 10, 0, 10])
plt.vlines([4], 0, 10, linestyles='dashed', lw=1.)
plt.hlines([5], 0, 4, linestyles='dashed', lw=1.)
plt.hlines([6.5], 4, 10, linestyles='dashed', lw=1.)
plt.vlines([2], 0, 5, linestyles='dashed', lw=1.)
plt.vlines([2.5], 5, 10, linestyles='dashed', lw=1.)
plt.vlines([6], 0, 6.5, linestyles='dashed', lw=1.)
plt.hlines([2], 0, 2, linestyles='dashed', lw=1.)
plt.hlines([9], 0, 2.5, linestyles='dashed', lw=1.)

plt.text(4.3, 7.3, 'A (4, 7)', fontsize=16)
plt.text(2.2, 4.5, 'B (3, 5)', fontsize=16)
plt.text(7, 6.8, 'C (8. 6.5)', fontsize=16)
plt.text(2.1, 1.3, 'D (2, 1)', fontsize=16)
plt.text(0.3, 6.3, 'E (2.5, 6)', fontsize=16)
plt.text(6.3, 3.3, 'F (6, 3)', fontsize=16)
plt.text(0.1, 2.3, 'G (1, 2)', fontsize=16)
plt.text(0.3, 9.3, 'H (1.5, 9)', fontsize=16)

plt.tick_params(labelbottom="off", left="off")
plt.tick_params(labelleft="off", left="off")
plt.show()
```
