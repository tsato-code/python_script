```python
import numpy as np
import matplotlib.pyplot as plt


def proj_point_to_line(p, seg_bgn, seg_end):
    a = [seg_end[0] - seg_bgn[0], seg_end[1] - seg_bgn[1]]
    b = [p[0] - seg_bgn[0], p[1] - seg_bgn[1]]
    k = np.dot(a, b)/np.dot(a, a)
    # k = np.absolute(k)
    ka = [k*a[0]+seg_bgn[0], k*a[1]+seg_bgn[1]]
    return ka


def check_angle(p, seg_bgn, seg_end):
    a = [seg_bgn[0] - p[0], seg_bgn[1] - p[1]]
    b = [seg_bgn[0] - seg_end[0], seg_bgn[1] - seg_end[1]]
    c = [seg_end[0] - p[0], seg_end[1] - p[1]]
    is_sharp = np.dot(a, b) >= 0 and -np.dot(b, c) >= 0
    return is_sharp
```

```python
plt.figure(figsize=(5,5))
plt.xlim([-10,10])
plt.ylim([-10,10])
plt.grid(which='major',color='black',linestyle='--')
for _ in range(10):
    data = [[np.random.randint(-10, 10), np.random.randint(-10, 10)] for _ in range(3)]     
    if check_angle(*data):
    # if True:
        proj_p = proj_point_to_line(*data)
        plot_y = [data[1][0], data[2][0]]
        plot_x = [data[1][1], data[2][1]]
        plot_p = [[data[0][0], proj_p[0]], [data[0][1], proj_p[1]]]
        plt.plot(plot_y, plot_x, '-', color='#1f77b4')
        plt.plot(*plot_p, '--o', color='#ff7f0e',alpha=0.9)
plt.show()
```
