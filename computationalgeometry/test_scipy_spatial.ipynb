```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Voronoi, voronoi_plot_2d
from scipy.spatial import Delaunay, delaunay_plot_2d
```

```python
n = 10
points = np.random.rand(10, 2)
```

```python
vor = Voronoi(points)
voronoi_plot_2d(vor)
plt.show()
```

```python
tri = Delaunay(points)
delaunay_plot_2d(tri)
plt.show()
```
