```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
times = 50
num = 12
```

```python
# 
X = np.random.normal(0, 1, size=(num, times))

# min, max 
upper = np.max(X)
lower = np.min(X)

# 
cols = np.ceil(num**0.5)
rows = np.ceil(num/cols)

# 
axes = []
fig = plt.figure(figsize=(16, 8))
for i, x in enumerate(X):
    axes.append(fig.add_subplot(rows, cols, i+1))
    axes[i].plot(x)
    
    # 
    if (i+1)//rows>=rows:
        axes[i].set_xlim([0, times])
    else:
        axes[i].tick_params(labelbottom="off")
    if i%cols==0:
        axes[i].set_ylim([lower, upper])
    else:
        axes[i].tick_params(labelleft="off")
```
