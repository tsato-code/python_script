```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.manifold import TSNE

def tsne_show(skdataset):
    skdataset_reduced = TSNE(n_components=2, random_state=0).fit_transform(skdataset.data)
    plt.scatter(skdataset_reduced[:, 0], skdataset_reduced[:, 1], c=skdataset.target, alpha=.5)
    plt.colorbar()
    plt.show()
```

```python
digits = datasets.load_digits()
tsne_show(digits)
```

```python
cancer = datasets.load_breast_cancer()
tsne_show(cancer)
```

```python
diabetes = datasets.load_diabetes()
tsne_show(diabetes)
```
