```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn import discriminant_analysis
```

```python
digits = datasets.load_digits()
digits.data.shape
```

```python
lda = discriminant_analysis.LinearDiscriminantAnalysis(n_components=2,solver='svd')
lda.fit(digits.data, digits.target)
```

```python
X_transformed = lda.fit_transform(digits.data, digits.target)
```

```python
plt.figure()
plt.scatter(X_transformed[:, 0], X_transformed[:, 1], c=digits.target, alpha=.5)
plt.colorbar()
plt.show()
```
