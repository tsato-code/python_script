## 2章の1変数ホテリング$T^2$法

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
```

```python
# Boston House Prices dataset
boston = datasets.load_boston()
plt.hist(boston.target)
plt.show()
```

```python
# 訓練データとテストデータに分割
y_train, y_test = train_test_split(boston.target, test_size=0.1, random_state=42)

# 平均と標本分散の計算
mu = np.mean(y_train)
s2 = np.var(y_train)

# カイ2乗分布による1%水準
th = sp.stats.chi2.ppf(0.99, 1)
```

```python
# 訓練データ
# 異常度の計算
a_train = (y_train - mu)**2/s2

plt.scatter(range(len(a_train)), a_train)
plt.hlines(th, 0, len(a_train), "r", linestyles='dashed')
plt.show()
```

```python
# テストデータ
# 異常度の計算
a_test = (y_test - mu)**2/s2

plt.scatter(range(len(a_test)), a_test)
plt.hlines(th, 0, len(a_test), "r", linestyles='dashed')
plt.show()
```
