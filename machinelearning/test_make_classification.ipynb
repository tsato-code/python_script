```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
# n_samples 生成するサンプル数 default=100
# n_features 特徴量の数 default=20
# n_informative 目的変数のラベルと相関が強い特徴量 default=2
# n_redundant Informative featureの線形結合から作られる特徴量の数 default=2
# n_repeated Infomative、Redundant featureのコピーからなる特長量の数 default=0
# n_classes 分類クラス数 default=2
# n_clusters_per_class 1クラスあたりのクラスタ数 default=2
# weights クラス内サンプル数の比率 default=None
# flip_y クラスのフリップ率で、0.01とすると各クラスの0.01%のサンプルの符号がランダムに変更される default=0.01
# hypercube 生成器に関するパラメータで、 True にすると超立方体の頂点にクラスタを配置 default=True
# random_state 乱数を制御するパラメータ default=None
```

```python
# 2次元のデータセット
dataset = make_classification(n_features=2, n_informative=2, n_redundant=0, n_repeated=0, flip_y=.2,
                              random_state=0, weights=[0.5, 0.2], n_clusters_per_class=1)
X, y = dataset
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k')
plt.show()
```

```python
# 1次元のデータセット
dataset = make_classification(
    n_features=1, n_informative=1, n_redundant=0, n_repeated=0,
    random_state=0, weights=[0.5, 0.2], n_clusters_per_class=1
)
X, y = dataset
plt.hist(X)
plt.show()
```
