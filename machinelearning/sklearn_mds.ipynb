```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn import manifold
```

```python
digits = datasets.load_digits()
digits.data.shape
```

```python
mds = manifold.MDS(n_components=2, random_state=0)
mds.fit(digits.data)
```

```python
X_transformed = mds.fit_transform(digits.data[:300])
```

```python
plt.figure()
plt.scatter(X_transformed[:, 0], X_transformed[:, 1], c=digits.target[:300], alpha=1.)
plt.colorbar()
plt.show()
```
