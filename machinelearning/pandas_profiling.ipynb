```python
# conda install -c conda-forge category_encoders
import category_encoders as ce
import pandas as pd
import matplotlib.pyplot as plt
```

```python
input_data_file = "../data/mushroom.csv"
```

```python
df = pd.read_csv(input_data_file, index_col=None, header=None)
print(df.head(3))
print("size:", df.shape)
print(df.describe())
```

```python
plt.figure()
df[0].value_counts().plot(kind='bar')
df[1].value_counts().plot(kind='bar')
plt.show()
```

```python
for i in range(1, len(df.columns)):
    pd.crosstab(columns=df[0], index=df[i], rownames=["df[{}]".format(i)], colnames=["df[0]"]).plot(kind="bar", title="(0, {})".format(i), figsize=(4, 3));
    plt.show()
```
